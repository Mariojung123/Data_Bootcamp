```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# 엑셀 파일 경로 설정
data_modify = '화장품_전처리.xlsx'

# 엑셀 파일을 데이터프레임으로 불러오기
df = pd.read_excel(data_modify)
```

```python
df['피부타입'] = df['피부타입'].str.replace('\n', ' ')
```

```python
df['피부고민'] = df['피부고민'].str.replace('\n', ' ')
```

```python
df['자극도'] = df['자극도'].str.replace('\n', ' ')
```

```python
df
```

```python
tfidf=TfidfVectorizer()

tfidf_matrix= tfidf.fit_transform(df['피부정보']).toarray()
tfidf_matrix
```

```python
output_excel_path = '화장품_전처리.xlsx'
```

```python
df.to_excel(output_excel_path, index=False)
```
