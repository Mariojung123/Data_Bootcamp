```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
```

```python
title_list = list()
price_list = list()
area_list = list()

driver = webdriver.Chrome()

driver.get('https://www.golmarket.co.kr/used_list.php')

search_box = driver.find_element(By.NAME, 'stx')

search_box.send_keys('스파이더')
search_box.send_keys(Keys.RETURN)

# 결과 페이지가 로드될 때까지 잠시 기다립니다.
driver.implicitly_wait(3)

# 정보를 담은 박스를 선택합니다.
search_results_title = driver.find_elements(By.CSS_SELECTOR, 'div.wr_subject')
search_results_price = driver.find_elements(By.CSS_SELECTOR, 'span.price')
search_results_area = driver.find_elements(By.CSS_SELECTOR, 'span.area')


for result in search_results_title:
    title_list.append(result.text)
    
for result in search_results_price:
    price_list.append(result.text)
    
for result in search_results_area:
    area_list.append(result.text)

    
# 작업이 끝나면 WebDriver를 종료하여 브라우저 창을 닫습니다.
driver.quit()
```

```python
for i in range(len(price_list)):
    if '원' in price_list[i]:
        price_list[i] = price_list[i].replace('원', '')
```

```python
for i in range(len(price_list)):
    price_list[i] = int(price_list[i].replace(",",""))
```

```python
golf_item_df = pd.DataFrame([title_list, price_list, area_list]).T
golf_item_df.columns = ['상품명', '판매가', '지역']
```

```python
golf_item_df.drop(golf_item_df[golf_item_df['판매가'] < 10000].index, inplace=True)
```
