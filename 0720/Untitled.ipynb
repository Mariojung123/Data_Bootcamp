```python
#1. 필요한 라이브러리 import 하기 (10점)

from bs4 import BeautifulSoup
import requests
import re
```

```python
#2. 링크를 request.get으로 가져오기 (header가 필요한 경우 headers 입력) (10점)

res = requests.get("https://www.golmarket.co.kr/used_list.php")


#3. BeautifulSoup을 이용해 파싱하기 (10점)

soup = BeautifulSoup(res.content, 'html.parser')


```

```python
#4. 상품명 크롤링하여 리스트로 저장하기 (20점)

golf_name = soup.select('.jongitem > .wr_subject')
golf_price = soup.select('.price > b')
golf_date = soup.select('.date')
golf_area = soup.select('.area')
```

```python
golf_list_name = list()
golf_list_price = list()
golf_list_date = list()
golf_list_area = list()
```

```python
for name in golf_name:
  golf_list_name.append(name.get_text())

for price in golf_price:
  golf_list_price.append(price.get_text())

for date in golf_date:
  golf_list_date.append(date.get_text())

for area in golf_area:
  golf_list_area.append(area.get_text())
```

```python
import pandas as pd

golmarket_df = pd.DataFrame([golf_list_name, golf_list_price, golf_list_date, golf_list_area]).T

golmarket_df.columns = ['상품명', '가격', '등록일', '위치']
```
